```python
import matplotlib.pyplot as plt
import csv
import numpy as np
f=open('ta_20220108214343.csv')
data=csv.reader(f)

next(data)

avg_temp=[]

for row in data:
    if row[2]!='':
        avg_temp.append(float(row[2]))

plt.figure(figsize = (8, 3))
plt.plot(avg_temp)

plt.rc('font', family = 'Malgun Gothic')
plt.title('부산의 평균 온도 추이')
plt.xlabel('1907년부터 2020년까지 온도 추이')
plt.ylabel('온도(℃)')
plt.show()
```
